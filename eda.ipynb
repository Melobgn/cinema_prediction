```python
# Calculate the z-score for each variable
z = np.abs(stats.zscore(dataframe['duree']))

# Identify outliers as students with a z-score greater than 3
threshold = 3
outliers = dataframe[z > threshold]

# Print the outliers
print(outliers)
```

```python
#Corrélation entre toutes les colonnes
plt.figure(figsize=(14,7))
corr = dataframe.corr(numeric_only=True)
matrix = np.triu(corr)
sns.heatmap(corr, cmap="RdBu_r", annot=True, mask=matrix, vmin=-1, center=0, vmax=1)
plt.show()
```

```python
topstudiocount = dataframe['Est_Top_Studio'].value_counts()
labels = ["0", "1"]
colors = ["orchid", "cornflowerblue"]

plt.pie(topstudiocount, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor':'white'})
plt.title('Popular studio or not')
plt.show()
```
