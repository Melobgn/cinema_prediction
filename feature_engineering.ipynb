```python
# convertir les noms des acteurs en un ensemble pour une recherche rapide
ensemble_acteurs = set(acteurs_df['name'])

def compter_acteurs_realisateurs(liste_acteurs, realisateur):
    total_correspondances = 0
    # Vérifier si la liste des acteurs est une chaîne de caractères et la traiter
    if isinstance(liste_acteurs, str):
        acteurs = liste_acteurs.split(',')
        total_correspondances += sum(acteur.strip() in ensemble_acteurs for acteur in acteurs)
    
    # Ajouter le réalisateur à la comptabilisation s'il est dans la liste
    if isinstance(realisateur, str) and realisateur.strip() in ensemble_acteurs:
        total_correspondances += 1
    
    return total_correspondances

# Appliquer cette fonction modifiée en utilisant apply avec un lambda pour passer les deux colonnes
df['Nb_celebrites'] = df.apply(lambda row: compter_acteurs_realisateurs(row['acteurs'], row['realisateur']), axis=1)

# Vérifier à nouveau le résultat
df.head(20)
```

```python
import pandas as pd


mean_entries = df.groupby('studio')['entrees'].mean()

# seuils pour les catégories
seuils = {
    'très grand studio': 250000,
    'grand studio': 100000,
    'studio normal': 20000,
    # studios en dessous de 20000 seront classés comme 'studio moyen' par défaut
}

# fonction pour catégoriser les studios en fonction des seuils d'entrée
def categoriser_studio(entrees):
    for categorie, seuil in seuils.items():
        if entrees >= seuil:
            return categorie
    return 'studio moyen'

#  nouvelle colonne 'categorie_studio' dans le DataFrame mean_entries
mean_entries = mean_entries.to_frame('entrees')  # Convertir la série en DataFrame pour ajouter une colonne
mean_entries['categorie_studio'] = mean_entries['entrees'].apply(categoriser_studio)

#ajoutez la catégorie de chaque studio au DataFrame original
#dictionnaire des moyennes et des catégories pour chaque studio
studio_categories = mean_entries['categorie_studio'].to_dict()

# map des catégories sur le DataFrame original
df['categorie_studio'] = df['studio'].map(studio_categories)

df.head(20)
```

### Créer un poids sur les acteurs

```python
#normalisation à la main (sur le même principe que minmaxscaler)
actors = pd.read_csv('acteurs.csv')

actors['coef_poids'] = (actors['prix'] - actors['prix'].min()) / (actors['prix'].max() - actors['prix'].min())

print(actors)
actors['coef_poids'].max()
actors.tail(10)
```

```python
# Trouver l'indice de l'élément maximum dans la colonne 'coef_poids'
indice_max_coef_poids = actors['coef_poids'].idxmax()

# Obtenir l'acteur correspondant à cet indice
acteur_max_coef_poids = actors.loc[indice_max_coef_poids, 'name']

print("L'acteur avec le coefficient de poids le plus élevé est:", acteur_max_coef_poids)
```
